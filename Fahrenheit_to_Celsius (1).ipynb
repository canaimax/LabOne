```
import tensorflow as tf
import numpy as np
```

```
fahrenheit = np.array([-22, 5, 32, 59, 86, 122, 212], dtype=float)
celsius    = np.array([-30, -15, 0, 15, 30, 50, 100], dtype=float)

for i,c in enumerate(celsius):
  print("{} degrees Fahrenheit = {} degree Celsius".format(fahrenheit[i], c))
 
```

```
def create_model():
  # Create a Dense layer l0 that contains only one neuron
  l0 = tf.keras.layers.Dense(units=1, input_shape=[1])

  # Create a sequential model with l0 layer
  model = tf.keras.Sequential([l0])

  # Choose a loss function and optimizer for the model
  model.compile(loss='mean_squared_error',
                optimizer=tf.keras.optimizers.Adam(0.1))
  
  return model
```

```
import datetime
# Create a log folder and a child folder named by the current time
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

# Create a tensorboard object to store the log
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir,
                                                      histogram_freq=1)

# Training the Fahrenheit to Celsius model
model = create_model()
model.fit(fahrenheit, celsius, epochs=500, verbose=False,
          callbacks=[tensorboard_callback])
print("Finished training the model")

```
